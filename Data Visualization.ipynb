```python
import matplotlib.pyplot as plt
#Array, list, Tupples and pandas series
import seaborn as sns
#DataFrame
import numpy as np
import pandas as pd
```

```python
year = [2015,2016,2017,2018,2019]
sales = [1234,4344,4454,3222,3343]
profit = [223,334,352,333,434]
```

```python
plt.plot(year,sales)
```

```python
plt.plot(year,sales,'red')
plt.xlabel('Year')
plt.ylabel('Sales and Profit')
plt.plot(year,profit,'blue')
plt.title('Sales and Profit by Year')
plt.show()
```

```python
plt.axes([0.05,0.05,0.9,0.9])
plt.plot(year,sales,'red')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Vs Year')
plt.axes([1.1,0.05,0.9,0.9])
plt.plot(year,profit,'blue')
plt.xlabel('Year')
plt.ylabel('Profit')
plt.title('Profir Vs Sales')
```

```python
plt.subplot(2,1,1)
plt.plot(year,sales,'red')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Vs Year')
plt.subplot(2,1,2)
plt.plot(year,profit,'blue')
plt.xlabel('Year')
plt.ylabel('Profit')
plt.title('Profir Vs Sales')
plt.tight_layout()
```

```python
plt.plot(year,sales,'red')
plt.xlim(2017,2018)
plt.ylim(4000,4500)
```

```python
plt.plot(year,sales,'brown')
plt.axis((2017,2018,4000,4500))
```

```python
plt.scatter(x=sales,y=profit,marker='+',label='sales vs profit')
plt.legend(loc=0)
plt.annotate('customer1',xy=(3343,333))
plt.style.use('seaborn-darkgrid')
```
